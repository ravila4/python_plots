# Plotting Error Bars

Here, we call matplotlib's ```errorbar``` function to display data with a continuous error measurement.

````python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# Linspace returns evenly spaced numbers over a specified interval.
x = np.linspace(0, 2 * np.pi)
# Defining the functions
y_sin = np.sin(x)
y_cos = np.cos(x)

# Create errorbar plot with a continuous error of ±0.2
plt.errorbar(x, y_sin, 0.2)
plt.errorbar(x, y_cos, 0.2)

plt.show()
````

If the error bars become distracting, another alternative is to use matplotlib's ```fillbetween``` function to draw two extra curves and shade the area between them as the error. Here we define a funcion that does this.

````python
def errorfill(x, y, yerr, color, alpha_fill=0.3, ax = None):
            
    # Define error range
    ymin = y - yerr
    ymax = y + yerr
    
    # Plot the original function
    plt.plot(x, y, color = color)
    
    # Plot the error range, and shade it with the same color
    plt.fill_between(x, ymax, ymin, color = color, alpha=alpha_fill)
````

````python
# Plotting sine and cosine functions with an error of ±0.2
errorfill(x, y_sin, 0.2, color = "red")
errorfill(x, y_cos, 0.2, color = "blue")

plt.show()
````
